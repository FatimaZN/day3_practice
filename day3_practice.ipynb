## Part 2 · Arithmetic Operators ➕


**2.1** — Set `a = 17`, `b = 5`. Print `a // b`, `a % b`, and `a ** 2` using f-strings.

```python
a = 17 
b = 5 
print(f"{a} // {b} = {a // b}")
print(f"{a} % {b} = {a % b}")
print(f"{a} ** 2 = {a**2}")
```

**2.2** — Ask the user for a number and print whether it's even — by printing the **remainder** when divided by 2 (`% 2`). *(Just print the remainder; no `if` yet.)*

## Part 3 · Assignment Operators 📥


**3.1** — Start `coins = 100`. Use `+=` to add 50, then `-=` to spend 30, then `*=` to double. Print after each step.

```python
coins = 100
coins += 50
print(f"coins1 += 50: {coins}")
coins -= 30
print(f"coins -= 30: {coins}")
coins *= 2
print(f"coins *=2 : {coins} ")
```

## Part 4 · Comparison Operators ⚖️ (print the True/False)


**4.1** — Print the result of: `15 == 15`, `8 != 8`, `100 > 99`, `3 <= 2`.

```python
print(f"{15 == 15}")
print(f"{8 != 8}")
print(f" {100>99}")
print(f"{3 <= 2}")
```

**4.2** — Ask the user for their age, then print `True/False` for: *is the age greater than 18?* (just print the comparison result).

```python
age = int(input("What is your age"))
print(f"Is the age greater than 18?  : {age >= 18}")
```

## Part 5 · Logical Operators 🔗 (and / or / not)


**5.1** — Print the result of each:
- `(10 > 5) and (2 > 8)`
- `(10 > 5) or (2 > 8)`
- `not (1 == 1)`


```python
print(f" {(10 > 5) and (2 > 8)} ")
print(f" {(10 > 5) or (2 > 8)} ")
print(f" {not(1 == 1)}")
```

## Part 6 · Membership Operators 🔎 (in / not in)


**6.1** — Set `word = "Python"`. Print whether `"P"` is in it, `"z"` is in it, and `"x" not in` it.

```python
word = "Python"
print(f"Is P in the word? -> {'P' in word}")
print(f"Is z in the word? -> {'z' in word}")
print(f"'x' not in word -> {'x' not in word}")
```
